## **Data Cleaning and Filtering**
<a href=#cont>Back to Table of Contents</a>

* **Purpose:** The purpose of this section is to prepare the data for analysis by addressing any issues such as missing values, outliers, and errors. Additionally, this step may involve filtering the data to focus on relevant features or rows to ensure accurate and meaningful analysis.
  
* **Details:** Handling Missing Values: Missing values in a dataset can cause issues in analysis. We will check for any missing values and decide whether to impute, drop, or handle them using other strategies (e.g., forward fill, mean/mode imputation).

* **Removing Outliers:** Outliers can skew the results of the analysis. We will identify and remove extreme values based on statistical methods (e.g., Z-score or IQR).

* **Correcting Errors:** Sometimes, the dataset may contain incorrect or inconsistent values. We will inspect for such errors and fix them where necessary.

* **Filtering Data:** In some cases, not all columns or rows are necessary for analysis. We may filter out irrelevant data to reduce noise and focus on the key insights.
---

```python
# Convert columns that should be numeric
df = df.apply(pd.to_numeric, errors='coerce')

# Check for missing values after conversion (if any invalid strings were converted to NaN)
print(df.isnull().sum())
```

```python
#Please use code cells to code in and do not forget to comment your code.
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Check for outliers using IQR (Interquartile Range) method on numeric columns
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Filter the dataset by removing outliers
df_filtered = numeric_df[~((numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))).any(axis=1)]

# Display the cleaned and filtered dataset
df_filtered.head()
```

## **HeatMap of Vegetable Price Correlations**

A heatmap is a graphical representation of data where individual values are depicted by color. It is primarily used to visualize the strength of relationships between variables in a dataset, typically in the form of a correlation matrix.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

The heatmap illustrates the correlation coefficients between the prices of different vegetables in the dataset. Key insights include:

Strength of Relationships: Darker colors indicate stronger correlations, with values close to 1 or -1 showing a strong positive or negative relationship, respectively.

Positive Correlations: Several vegetables, such as Tomato and Onion, may exhibit strong positive correlations, suggesting that their prices tend to move together.

Weak or Negative Correlations: Some vegetables may show weak or negative correlations, indicating little to no relationship in their pricing trends.

Market Insights: Understanding these correlations can help stakeholders make informed decisions regarding pricing strategies, supply chain management, and product pairings.

Overall, the heatmap provides a clear visual representation of the interdependencies among vegetable prices, highlighting areas for potential analysis and strategic planning.

## **Histogram of Vegetable Prices**

A histogram is a type of bar chart that represents the distribution of a dataset by grouping data points into ranges or intervals, called bins. It helps to visualize how frequently data values occur within these bins.

```python
import matplotlib.pyplot as plt

df.hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

The histogram displays the distribution of prices for the different vegetables in the dataset. Key observations include:

Price Distribution: The histograms show the spread of prices for each vegetable, revealing which price ranges are most common.

Skewness: Some vegetables may exhibit skewness, indicating that prices are concentrated towards lower or higher values. For example, if the histogram for tomatoes shows a peak on the lower end, it suggests most prices are relatively low.

Outliers: Any long tails in the distribution may indicate the presence of outlier prices that are significantly higher or lower than the rest.

Overall, the histogram provides valuable insights into the pricing patterns of the vegetables, which can aid in understanding market dynamics and pricing strategies.

## *Tomato vs. Onion Scatter Plot*

```python
# Scatter plot for 'Tomato' vs 'Onion'
plt.figure(figsize=(10, 6))
plt.scatter(df['Tomato'], df['Onion'], color='b')

plt.title('Scatter Plot of Tomato vs Onion Prices')
plt.xlabel('Tomato Prices')
plt.ylabel('Onion Prices')
plt.grid(True)
plt.show()
```

The scatter plot illustrates the relationship between Tomato and Onion prices. Key observations include:

Positive Correlation: There appears to be a positive correlation, indicating that as the price of tomatoes increases, the price of onions tends to rise as well.

Clustered Data Points: Most points are clustered along a diagonal line, suggesting a consistent pricing trend between the two vegetables.

Potential Outliers: A few points may lie outside the main cluster, indicating unusual pricing scenarios that could be explored further.

Overall, this scatter plot suggests a strong relationship between tomato and onion prices, which may influence market strategies for both vegetables.

## *Scatter plots of the relationships between vegetable prices*

```python
import itertools
import math

# List of vegetable columns
vegetables = ['Bhindi (Ladies finger)', 'Tomato', 'Onion', 'Potato', 
              'Brinjal', 'Garlic', 'Peas', 'Methi', 'Green Chilli', 
              'Elephant Yam (Suran)']

# Create scatter plots for each pair of vegetables
num_combinations = len(list(itertools.combinations(vegetables, 2)))  # Total combinations
cols = 3  # Number of columns for the subplot
rows = math.ceil(num_combinations / cols)  # Calculate number of rows

plt.figure(figsize=(15, 5 * rows))

# Generate scatter plots for all combinations of vegetable pairs
for i, (veg1, veg2) in enumerate(itertools.combinations(vegetables, 2)):
    plt.subplot(rows, cols, i + 1)  # Adjust the layout according to the number of plots
    plt.scatter(df[veg1], df[veg2], color='b', alpha=0.7)
    plt.title(f'{veg1} vs {veg2}')
    plt.xlabel(veg1)
    plt.ylabel(veg2)
    plt.grid(True)

plt.tight_layout()
plt.show()
```

The scatter plots reveal the relationships between vegetable prices, highlighting:

Positive Correlations: Certain vegetables, like Bhindi (Ladies finger) and Tomato, show a tendency to have rising prices together, indicating similar market trends.

Negative Correlations: Some pairs, such as Onion and Potato, may exhibit inverse relationships, suggesting they could be substitutes in consumer preferences.

Outliers: Isolated points far from the trend indicate unusual pricing, possibly due to market fluctuations or seasonal changes.

Market Insights: The correlations observed can help stakeholders make informed decisions about pricing and inventory management.

Overall, these scatter plots provide valuable insights into the dynamics of vegetable pricing and market behavior.
